```python
#Import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
#Create empty set with the necessary names
xy = ['X','Y']
#Read the data file seprated by comma using the names given above
df = pd.read_csv('challenge_dataset.txt',sep=',',names='xy')
#Substitiut X_values and Y_values to the necessary data sets
X_values = df[['x']]
Y_values = df[['y']]
#Proceed to apply the linear model using linear regression
body_reg = linear_model.LinearRegression()
#Fit the given linear model
body_reg.fit(X_values, Y_values)
#Then scatter the given plot based on the data
plt.scatter(X_values,Y_values)
#Predict the given values based the plotted data
plt.plot(X_values,body_reg.predict(X_values))
#Show the given result
plt.show()
```
